```python
import matplotlib.pyplot as plt

apl_price = [93.95, 112.15, 104.05, 144.05, 144.85, 169.49]
org_price = [39.01, 50.29, 29.57, 57.05, 69.68, 94.39]
year = [2014, 2015, 2016, 2017, 2018, 2019]  # Added 2019 to match the length

plt.plot(year, apl_price, marker='o', label='APL Price')
plt.plot(year, org_price, marker='o', label='Organization Price')

plt.xlabel('Year')
plt.ylabel('Price')
plt.title('APL and Organization Prices over the Years')
plt.grid(True)
plt.legend()
plt.show()

```
